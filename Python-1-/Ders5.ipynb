```python
# Beden kitle indexini hesaplayan bir fonksiyon yaziniz.
# 2 arguman aliyor, boy, kilo
# fonksiyon boy ve kilo ile calistiginda beden kitle indexini verecek mesaj vericek, uyari ya da tebrikler
# Vucut kitle indeksi (VKI) => Vucut agirligi(kg)/BoyxBoy

def vki(kilo, boy):
    vki = kilo/(boy/100) ** 2
    if vki > 0 and vki < 18.5:
        return f"Uyari! Vucut kitle indeksiniz {vki:.2f} olarak hesaplandi. Zayif"
    elif vki >= 18.5 and vki < 24.9:
        return f"Tebrikler! Vucut kitle indeksiniz {vki:.2f} olarak hesaplandi. Normal"
    elif vki >= 24.9 and vki < 29.9:
        return f"Uyari! Vucut kitle indeksiniz {vki:.2f} olarak hesaplandi.  Immhhh tam emirin agzina layiksiniz <3"
    elif vki >= 29.9:
        return f"Uyari! Vucut kitle indeksiniz {vki:.2f} olarak hesaplandi. Obezite"
    else:
        return "Hatali deger girdiniz."
    
vki(68, 164)
```

```python
num1 = 5
num2 = 'string'

try:
    num1 / num2
except ZeroDivisionError as e:
    print('Sifir ile bolme mumkun degil!')
    print(e)
except TypeError as e:
    print('Data tipinde bir sorun var!')
    print(e)
except:
    print('Birseyler yanlis gidiyor!')
```

DECORATORS

```python
# * Decorators nedir?
# Python'da decorator, bir fonksiyonun davranışını değiştiren bir fonksiyondur. Decorator'lar, fonksiyonların önüne eklenir ve fonksiyon çağrıldığında çalıştırılırlar. Decorator'lar, fonksiyonlara yeni özellikler eklemek veya fonksiyonların davranışını değiştirmek için kullanılır.

# Decorator'lar, Python'da çok çeşitli amaçlar için kullanılabilir. Örneğin, decorator'lar şunlar için kullanılabilir:

# * Fonksiyonların önbelleğe alınmasını sağlamak
# * Fonksiyonların zamanlamasını ölçmek
# * Fonksiyonlara yetkilendirme eklemek
# * Fonksiyonların hata ayıklamasını kolaylaştırmak

# Decorator'lar, Python'da çok güçlü bir araçtır ve birçok farklı amaç için kullanılabilirler.

def my_decorator(func):
    def wrapper():
        # Do something before
        result = func()
        # Do something after
        return result  # or erase this line, with only func() above
    return wrapper


@my_decorator
def func():
    pass
```

```python
def my_decorator(func):
    def wrapper():
        print('----------------') # Do something before
        result = func() # or only func()
        print(result)
        print('----------------') # Do something after
        # return result  # or erase this line, with only func() above
    return wrapper


@my_decorator
def func():
    return 'Hello World!'

func()
```

```python
import time

def gecen_zaman(func):
    def wrapper():
        # Do something before
        t1 = time.time()
        func()
        # Do something after
        t2 = time.time()
        return f'Gecen zaman: {((t2 - t1) * 1000):.2f} ms'  # or erase this line, with only func() above
    return wrapper


@gecen_zaman
def big_sum():
    num_list = []
    for num in (range(0, 1000)):
        num_list.append(num)
        print(f'Big sum: {sum(num_list)}')

big_sum()
```
