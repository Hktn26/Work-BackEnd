```python
class Person:

    company = 'EmHaCo'

    def __init__(self, isim, yas, cinsiyet):
        self.isim = isim
        self.yas = yas
        self.cinsiyet = cinsiyet
        print('Personel Olusturuldu')

    def __str__(self):
        return f'{self.isim} - {self.yas}'

    def get_detay(self):
        return f"Adi:{self.isim}, \nYas:{self.yas}, \nSirket:{self.company}, \nCinsiyet:{self.cinsiyet}"


# Bir classin atandigi degiskene instance denir
# Personel Olusturuldu __init__
person_2 = Person('Emir', 22, 'Erkek')
# Bir classtan turetilmis objeye instance denir.
# Emir - 22 __str__
print(person_2)
# get_detay ciktisi Adi:Emir, Yas:22, Sirket:EmHaCo, Cinsiyet:Erkek
print(person_2.get_detay())


class Department:

    def __init__(self):
        print('Departman calisit.')

    def set_departman(self, departman):
        self.departman = departman

# Inheritance:
# JS -> class Employee extends Person


class Employee(Person, Department):
    

    def set_maas(self, maas):
        self.maas = maas
        # Overriding metodunu kullanarak birden fazla calistirmak icin super() kullanin

    # Override:
    def __init__(self, isim, yas, cinsiyet, maas, departman):
        # Super(), inherit ettigimiz ilk class'dan itibaren buldugu ilk methodu cagirir.
        super().__init__(isim, yas, cinsiyet)
        # Inherit edilen class methodunu calistirma:
        # super() kullanmiyor isek self parametresi gosterilmelidir.
        Department.set_departman(self, departman)
        self.departman = departman
        self.maas = maas
    
    #Polymorphism (Override, Overload)
    #Mevcut bir methodu tekrar tanimlama imkanina 'Polymorphism' denir.
    #Yeni methodun eski methodu ezmesine 'Override' denir.
    #Farkli tipte parameter almaya izin veren 'Overloading' denir.
    #Yeni method ile eski methodun ayni anda aktif olmasina da 'Overload' denir.
    #Python 'Overload' desteklemez.

    def get_detay(self):
        return f"Adi:{self.isim}, \nYas:{self.yas}, \nSirket:{self.company}, \nCinsiyet:{self.cinsiyet}, \n{self.maas}, \n{self.departman}"

person_2 = Employee('Haktan', 23, 'Erkek', 2500, 'BackEnd')
print(person_2.__dict__)
print(person_2.get_detay())
```
